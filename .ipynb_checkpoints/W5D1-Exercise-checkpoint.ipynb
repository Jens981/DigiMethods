## 5.1.1 Data Wrangling and Visualizing

**Exercise 5.1.1.1** Read in the dataset "nchs_death_causes.csv" as a list of lists. You will need to use the `csv` package and then will want to go through each row appending it to a list you create.

```python
import csv

word_list = []

f = open ("nchs_death_causes.csv", "r", encoding='utf8')
reader = csv.reader(f)
for row in reader:
    word_list.append(row)
f.close()
```

**Exercise 5.1.1.2** Make a list that is a subset of the data, containing only one cause of death and one state.

```python
ny_liste = []

for item in word_list:
    if item[2]== "Alzheimer's disease" and item[3]== "Iowa":
        ny_liste.append(item)
```

```python
ny_liste
```

```python
list_of_tuples = [tuple(l) for l in ny_liste]
```

```python
list_of_tuples
```

**Exercise 5.1.1.3** Plot a line graph that shows the development of the death rate over time.

```python
list_of_tuples
```

```python
import matplotlib.pyplot as plt

def Sort_Tuple(tup):
    tup.sort(key = lambda x: x[0])
    return tup 

Sort_Tuple(list_of_tuples)

values = [item[4] for item in list_of_tuples]
labels = [item[0] for item in list_of_tuples]


ny_value = []

for i in values:
    ny_value.append(float(i))
    
ny_value
    


plt.scatter(labels, ny_value)
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()


```
