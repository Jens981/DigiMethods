```python
list_of_tuples = [("kim", 3), ("Ole", 8), ("Lars", 10), ("Simon", 20), ("WalleKirk", 15), ("Sine", 13)]
```

```python
def sort_list(tup):
    tup.sort(key = lambda x: x[1])  
    return tup 
```

```python
sort_list(list_of_tuples)
```

```python
values = [item[1] for item in list_of_tuples]
labels = [item[0] for item in list_of_tuples]
```

```python
import matplotlib.pyplot as plt

plt.bar(labels, values)
plt.xlabel("Navne")
plt.ylabel("Alder")
plt.title("Dette er en bargraf over aldre")
plt.tight_layout()
plt.show()
```
